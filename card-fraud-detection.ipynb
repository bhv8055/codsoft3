```python
fraud = card[card.Class == 1]
normal = card[card.Class == 0]
```

```python
card.groupby('Class').mean()
```

```python
fraud.Amount.describe()
```

```python
normal.Amount.describe()
```

```python
normal_sample = normal.sample(n=492)
```

```python
#sampling
new_d = pd.concat([normal_sample,fraud],axis=0)
```

```python
new_d.head()
```

```python
new_d['Class'].value_counts()
```

```python
new_d.groupby('Class').mean()
```

```python
#model building
x =new_d.drop(columns='Class',axis=1)
y =new_d['Class']
```

```python
x
```

```python
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=11)
```

```python
model = lm()
```

```python
model.fit(x_train,y_train)
```

```python
test_pred = model.predict(x_train)
train_acc = accuracy_score(test_pred,y_train)
```

```python
test_pred = model.predict(x_test)
test_acc = accuracy_score(test_pred,y_test)
```
